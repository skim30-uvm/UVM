```python
#Functions
from random import randrange

def display_board(board):
    # The function accepts one parameter containing the board's current status
    # and prints it out to the console.
    for i in range(13):
        if i%4 == 0:
            print(('+'+'-'*7)*3,end='+\n')
        elif i%4 == 1 or i%4 == 3:
            print(('|'+' '*7)*3,end='|\n')
        else:
            print('|'+' '*3+str(board[i//4][0])+' '*3,
                  '|'+' '*3+str(board[i//4][1])+' '*3,
                  '|'+' '*3+str(board[i//4][2])+' '*3,sep='',end='|\n')

def enter_move(board):
    # The function accepts the board current status, asks the user about their move, 
    # checks the input and updates the board according to the user's decision.
    while True:
        call = int(input("Please enter the number of spot for your move: ")) - 1
        if call<0 or call>8:
            print("Invalid move: put number 1~9")
            continue
        if isinstance(board[call//3][call%3],int):
            board[call//3][call%3] = 'O'
            break
        else:
            print("The place is occupied")
            continue
    return board

def make_list_of_free_fields(board):
    # The function browses the board and builds a list of all the free squares; 
    # the list consists of tuples, while each tuple is a pair of row and column numbers.
    lst = []
    for n in range(1,10):
        if isinstance(board[(n-1)//3][(n-1)%3],int):
            lst.append(((n-1)//3, (n-1)%3))
    return lst        

def victory_for(board, sign):
    # The function analyzes the board status in order to check if 
    # the player using 'O's or 'X's has won the game
    # - win
    if [sign,sign,sign] in board:
        return sign

    # | win
    for i in range(3):
        if [board[0][i],board[1][i],board[2][i]] == [sign,sign,sign]:
            return sign

    # / win
    if [board[0][0],board[1][1],board[2][2]] == [sign,sign,sign]\
        or [board[0][2],board[1][1],board[2][0]] == [sign,sign,sign]:
        return sign
    else:
        return


def draw_move(board):
    # The function draws the computer's move and updates the board.
    while True:
        put = randrange(8)
        if isinstance(board[put//3][put%3],int):
            board[put//3][put%3] = 'X'
            break
        else:
            continue
    return board

def draw_game(board):
    for n in range(1,10):
        if isinstance(board[(n-1)//3][(n-1)%3],int):
            return False
    return True
```

```python
#Initial borad
print("Let's start the game!")
board = [[1,2,3],[4,5,6],[7,8,9]]
turn = 0
print("["+"Turn "+str(turn)+"]",)
display_board(board)
print("Ready!")
```

```python
print("Computer turn")
board = [[1,2,3],[4,'X',6],[7,8,9]]
display_board(board)
while True:
    turn += 1
    print("Your turn!")
    board = enter_move(board)
    display_board(board)
    if victory_for(board, 'O') == 'O':
        print("Player win!")
        break
    if draw_game(board) == True:
        print("This game is a draw")
        break
    print("Computer turn")
    board = draw_move(board)
    display_board(board)
    if victory_for(board, 'X') == 'X':
        print("Computer win!")
        break
    if draw_game(board) == True:
        print("This game is a draw")
        break
print('Thank you for playing!')
```
